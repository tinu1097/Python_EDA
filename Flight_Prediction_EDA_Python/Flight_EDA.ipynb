```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pandas_profiling import ProfileReport
%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
#import train dataset
train_df = pd.read_excel("Data_Train.xlsx")
train_df.head(3)
```

```python
#import test dataset
test_df = pd.read_excel("Test_set.xlsx")
test_df.head(3)
```

```python
#merge train and test dataset
final_df =  train_df.append(test_df)
```

```python
#top 5 rows
final_df.head()
```

```python
#bottom 5 rows
final_df.tail()
```

```python
#information about dataset
final_df.info()
```

```python
#convert into datetime format
final_df["Date_of_Journey"] = pd.to_datetime(final_df["Date_of_Journey"])
```

```python
#create another three columns
final_df["Day"] = final_df['Date_of_Journey'].dt.day
final_df["Month"] = final_df['Date_of_Journey'].dt.month
final_df["Year"] = final_df['Date_of_Journey'].dt.year
```

```python
#drop 'Date_of_Journey' column
final_df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
#checking column names
final_df.columns
```

```python
#final_df["Arrival_Time"] = final_df["Arrival_Time"].str.split(" ").str[0]
```

```python
#convert datetime function
final_df["Arrival_Time"] = pd.to_datetime(final_df["Arrival_Time"])
final_df["Arrival_hr"] = final_df["Arrival_Time"].dt.hour
final_df["Arrival_min"] = final_df["Arrival_Time"].dt.minute
#drop column Arrival_Time 
final_df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
#show data frame for Arrival_hr,Arrival_min
final_df[["Arrival_hr","Arrival_min"]]
```

```python
final_df["Dep_Time"] = pd.to_datetime(final_df["Dep_Time"])
final_df["Dep_Time_hr"] = final_df["Dep_Time"].dt.hour
final_df["Dep_Time_min"] = final_df["Dep_Time"].dt.minute
final_df.drop("Dep_Time",axis=1,inplace=True)
```

```python
#check unique values
final_df["Total_Stops"].unique()
```

```python
#mapping the values 
final_df["Total_Stops"] = final_df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4})
```

```python
#counts the value
final_df["Total_Stops"].value_counts()
```

```python
#checking null values
final_df[final_df["Total_Stops"].isnull()]
```

```python
 #fill null values with mode value
 final_df["Total_Stops"] = final_df["Total_Stops"].fillna(final_df["Total_Stops"].mode()[0])
```

```python
#checking null values
final_df["Total_Stops"].isnull().sum()
```

```python
#drop Route column
final_df.drop(["Route"],axis=1,inplace=True)
```

```python
#split the duration column
final_df["Duration_hour"] = final_df["Duration"].str.split(" ").str[0].str.split("h").str[0]
```

```python
#checking where duration==5m
final_df[final_df["Duration_hour"]=="5m"]
```

```python
#drop rows where rows have 5m
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
#convert into int
final_df["Duration_hour"] = final_df["Duration_hour"].astype(int)
#multiply with 60
final_df["Duration_hour"] = final_df["Duration_hour"]*60
```

```python
#checking values
final_df["Duration_hour"]
```

```python
#split column with m
final_df["min"] = final_df["Duration"].str.split(" ").str[1].str.split("m").str[0]
```

```python
#same procedure apply with min
final_df["min"] = final_df["min"].fillna(0)
final_df["min"] = final_df["min"].astype(int)
final_df["Duration_min"] = final_df["Duration_hour"]+final_df["min"]
```

```python
#drop columns Duration_hour,min,Duration
final_df.drop(["Duration_hour","min","Duration"],axis=1,inplace=True)
```

```python
#checking column names
final_df[["Airline","Source","Destination"]]
```

```python
#import LabelEncoder 
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
#Lableling all columns 
final_df["Airline"] = labelencoder.fit_transform(final_df["Airline"])
final_df["Source"] = labelencoder.fit_transform(final_df["Source"])
final_df["Destination"] = labelencoder.fit_transform(final_df["Destination"])
final_df["Additional_Info"] = labelencoder.fit_transform(final_df["Additional_Info"])
```

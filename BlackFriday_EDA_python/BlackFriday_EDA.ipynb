```python
final_black['Gender'] = final_black['Gender'].map({"F":0,"M":1})
final_black["Age"] = final_black["Age"].map({"0-17":1,"18-25":2,"26-35":3,"36-45":4,"46-50":5,"51-55":6,"55+":7})
```

```python
#Second Technique for label encoding
#from sklearn import preprocessing
#label_encoder = preprocessing.LabelEncoder()
#final_black["Age"] = label_encoder.fit_transform(final_black["Age"])
#final_black["Age"]
```

```python
#find the unique values
final_black["City_Category"].unique()
```

```python
#use get_dummies for encoding
final_city = pd.get_dummies(final_black["City_Category"],drop_first=True,prefix="City")
```

```python
#concat final_city with final_black data frame
final_black = pd.concat([final_black,final_city],axis=1)
```

```python
#drop city_category column
final_black.drop(["City_Category"],inplace=True,axis=1)
```

```python
#checking null values
final_black.isnull().sum()
```

```python
#find value_counts
final_black["Product_Category_2"].value_counts()
```

```python
#replace missing vlaue with mod beacuse Product_Category_2  and Product_Category_3  its a ctegorcal feature
#we can get mode value by this
final_black["Product_Category_2"].mode()[0]
```

```python
final_black["Product_Category_2"] = final_black["Product_Category_2"].fillna(final_black["Product_Category_2"].mode()[0])
```

```python
final_black["Product_Category_2"].isnull().sum()
```

```python
final_black["Product_Category_3"].value_counts()
```

```python
final_black["Product_Category_3"] = final_black["Product_Category_3"].fillna(final_black["Product_Category_3"].mode()[0])
```

```python
final_black["Product_Category_3"].isnull().sum()
```

```python
final_black["Stay_In_Current_City_Years"].value_counts()
```

```python
final_black["Stay_In_Current_City_Years"]=final_black["Stay_In_Current_City_Years"].str.replace("+"," ")
```

```python
final_black["Stay_In_Current_City_Years"].value_counts()
```

```python
final_black.isnull().sum()
```

```python
#convert object into int64
final_black["Stay_In_Current_City_Years"]=final_black["Stay_In_Current_City_Years"].astype("int64")
final_black["City_B"]=final_black["City_B"].astype("int64")
final_black["City_C"]=final_black["City_C"].astype("int64")
```
